#### Brief description of the problem and data
Briefly describe the challenge problem and NLP. Describe the size, dimension, structure, etc., of the data.

The challenge problem revolves around using Generative Adversarial Networks (GANs) to create art in the style of Claude Monet. Participants are tasked with building a GAN architecture consisting of a generator and a discriminator. The generator aims to produce images resembling Monet's style, while the discriminator is trained to distinguish between real Monet paintings and generated ones.

The data provided for this task consists of TFRecord files containing images. There are 5 TFRecord files for Monet paintings and 20 for photographs, with respective sizes of 300 and 7038 images. The dimensions of the images in both datasets are uniform, with a size of 256x256 pixels and 3 color channels (RGB).

In summary, the challenge involves using GANs to generate Monet-style images, with the provided data consisting of a relatively small collection of Monet paintings and a larger collection of photographs, both standardized to a specific size and dimension.

```python
AUTOTUNE = tf.data.experimental.AUTOTUNE

MONET_FILENAMES = tf.io.gfile.glob('./monet_tfrec/*.tfrec')
print('Monet TFRecord Files:', len(MONET_FILENAMES))

PHOTO_FILENAMES = tf.io.gfile.glob('./photo_tfrec/*.tfrec')
print('Photo TFRecord Files:', len(PHOTO_FILENAMES))
```

```python
# Function to parse TFRecord files
def parse_tfrecord_fn(example):
    feature_description = {
        'image': tf.io.FixedLenFeature([], tf.string),
    }
    example = tf.io.parse_single_example(example, feature_description)
    image = tf.io.decode_jpeg(example['image'], channels=3) 
    return image

# Function to get dataset size
def get_dataset_size(filenames):
    num_samples = 0
    for filename in filenames:
        for record in tf.data.TFRecordDataset(filename):
            num_samples += 1
    return num_samples

# Function to get data dimensions
def get_data_dimensions(filenames):
    dimensions = set()
    for filename in filenames:
        for record in tf.data.TFRecordDataset(filename):
            example = tf.io.parse_single_example(record, {'image': tf.io.FixedLenFeature([], tf.string)})
            image = tf.io.decode_jpeg(example['image'], channels=3)
            dimensions.add(image.shape)
    return dimensions

# Get sizes and dimensions
monet_size = get_dataset_size(MONET_FILENAMES)
photo_size = get_dataset_size(PHOTO_FILENAMES)
monet_dimensions = get_data_dimensions(MONET_FILENAMES)
photo_dimensions = get_data_dimensions(PHOTO_FILENAMES)

# Print results
print("Monet dataset size:", monet_size)
print("Photo dataset size:", photo_size)
print("Monet data dimensions:", monet_dimensions)
print("Photo data dimensions:", photo_dimensions)
```

#### Exploratory Data Analysis (EDA) â€” Inspect, Visualize and Clean the Data
Show a few visualizations like histograms. Describe any data cleaning procedures. Based on your EDA, what is your plan of analysis?

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns

# Function to load dataset
def load_dataset(filenames):
    dataset = tf.data.TFRecordDataset(filenames)
    dataset = dataset.map(parse_tfrecord_fn, num_parallel_calls=tf.data.experimental.AUTOTUNE)
    return dataset

# Load Monet and photo datasets
monet_dataset = load_dataset(MONET_FILENAMES)
photo_dataset = load_dataset(PHOTO_FILENAMES)

# Convert datasets to NumPy arrays for visualization
def dataset_to_numpy(dataset, num_samples=5):
    images = []
    for image in dataset.take(num_samples):
        images.append(image.numpy())
    return images

monet_images = dataset_to_numpy(monet_dataset)
photo_images = dataset_to_numpy(photo_dataset)

# Plot histograms of pixel values for Monet and photo datasets
def plot_histograms(images, titles):
    fig, axs = plt.subplots(1, len(images), figsize=(15, 5))
    for i, (image, title) in enumerate(zip(images, titles)):
        flattened_image = image.flatten()
        sns.histplot(flattened_image, ax=axs[i], kde=True)
        axs[i].set_title(title)
        axs[i].set_xlabel('Pixel Value')
        axs[i].set_ylabel('Frequency')
    plt.tight_layout()
    plt.show()

plot_histograms(monet_images, titles=['Monet Image 1', 'Monet Image 2', 'Monet Image 3', 'Monet Image 4', 'Monet Image 5'])
plot_histograms(photo_images, titles=['Photo Image 1', 'Photo Image 2', 'Photo Image 3', 'Photo Image 4', 'Photo Image 5'])
```

```python
# Function to plot sample images
def plot_sample_images(images, titles):
    fig, axs = plt.subplots(1, len(images), figsize=(15, 5))
    for i, (image, title) in enumerate(zip(images, titles)):
        axs[i].imshow(image)
        axs[i].set_title(title)
        axs[i].axis('off')
    plt.tight_layout()
    plt.show()

# Get sample images from Monet and photo datasets
sample_monet_images = monet_images[:5]
sample_photo_images = photo_images[:5]

# Plot sample images
plot_sample_images(sample_monet_images, titles=['Monet Image 1', 'Monet Image 2', 'Monet Image 3', 'Monet Image 4', 'Monet Image 5'])
plot_sample_images(sample_photo_images, titles=['Photo Image 1', 'Photo Image 2', 'Photo Image 3', 'Photo Image 4', 'Photo Image 5'])
```

```python
# Function to extract RGB values from images
def extract_rgb(images):
    red_values = []
    green_values = []
    blue_values = []
    for image in images:
        red_values.extend(image[:,:,0].flatten())
        green_values.extend(image[:,:,1].flatten())
        blue_values.extend(image[:,:,2].flatten())
    return red_values, green_values, blue_values

# Extract RGB values from Monet and photo datasets
monet_red, monet_green, monet_blue = extract_rgb(monet_images)
photo_red, photo_green, photo_blue = extract_rgb(photo_images)

# Plot RGB scatter plots
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

# Monet dataset
axs[0].scatter(monet_red, monet_green, color='red', alpha=0.5, label='Red-Green')
axs[0].scatter(monet_red, monet_blue, color='blue', alpha=0.5, label='Red-Blue')
axs[0].scatter(monet_green, monet_blue, color='green', alpha=0.5, label='Green-Blue')
axs[0].set_title('Monet Dataset - RGB Scatter Plot')
axs[0].set_xlabel('Red')
axs[0].set_ylabel('Green / Blue')
axs[0].legend()

# Photo dataset
axs[1].scatter(photo_red, photo_green, color='red', alpha=0.5, label='Red-Green')
axs[1].scatter(photo_red, photo_blue, color='blue', alpha=0.5, label='Red-Blue')
axs[1].scatter(photo_green, photo_blue, color='green', alpha=0.5, label='Green-Blue')
axs[1].set_title('Photo Dataset - RGB Scatter Plot')
axs[1].set_xlabel('Red')
axs[1].set_ylabel('Green / Blue')
axs[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Function to preprocess images
def preprocess_image(image):
    # Resize image to 256x256 pixels
    image = tf.image.resize(image, [256, 256])
    # Normalize pixel values to range [0, 1]
    image = image / 255.0
    return image

# Function to load and preprocess dataset
def load_and_preprocess_dataset(filenames, batch_size):
    dataset = tf.data.TFRecordDataset(filenames)
    dataset = dataset.map(parse_tfrecord_fn, num_parallel_calls=tf.data.experimental.AUTOTUNE)
    dataset = dataset.map(preprocess_image, num_parallel_calls=tf.data.experimental.AUTOTUNE)
    dataset = dataset.batch(batch_size)
    return dataset

# Load and preprocess Monet dataset
monet_dataset = load_and_preprocess_dataset(MONET_FILENAMES, batch_size=32)
```

#### DModel Architecture
Describe your model architecture and reasoning for why you believe that specific architecture would be suitable for this problem. Compare multiple architectures and tune hyperparameters.

```python
def build_generator(latent_dim):
    inputs = layers.Input(shape=(latent_dim,))
    x = layers.Dense(256 * 16 * 16)(inputs)
    x = layers.Reshape((16, 16, 256))(x)
    x = layers.Conv2DTranspose(128, 4, strides=2, padding='same', activation='relu')(x)
    x = layers.Conv2DTranspose(64, 4, strides=2, padding='same', activation='relu')(x)
    x = layers.Conv2DTranspose(3, 4, strides=2, padding='same', activation='sigmoid')(x)  # Output image
    # Resize to 256x256
    x = tf.image.resize(x, [256, 256])
    return Model(inputs, x)
```

```python
def build_discriminator():
    inputs = layers.Input(shape=(256, 256, 3))
    x = layers.Conv2D(64, 4, strides=2, padding='same', activation='relu')(inputs)
    x = layers.Conv2D(128, 4, strides=2, padding='same', activation='relu')(x)
    x = layers.Conv2D(256, 4, strides=2, padding='same', activation='relu')(x)
    x = layers.Flatten()(x)
    outputs = layers.Dense(1, activation='sigmoid')(x)  # Binary classification (real/fake)
    return Model(inputs, outputs)

discriminator = build_discriminator()
discriminator.summary()
```

```python
def build_gan_model(learning_rate=0.0002, batch_size=32, latent_dim=100):
    # Build discriminator
    discriminator = build_discriminator()
    discriminator.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
                          loss='binary_crossentropy',
                          metrics=['accuracy'])

    # Make discriminator non-trainable during generator training
    discriminator.trainable = False

    # Build generator
    generator = build_generator(latent_dim=latent_dim)

    # GAN model
    gan_input = tf.keras.Input(shape=(latent_dim,))
    generated_image = generator(gan_input)
    gan_output = discriminator(generated_image)
    gan = tf.keras.Model(gan_input, gan_output)

    # Compile GAN
    gan.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
                loss='binary_crossentropy')

    return gan
```

#### Results and Analysis
Run hyperparameter tuning, try different architectures for comparison, apply techniques to improve training or performance, and discuss what helped.

Includes results with tables and figures. There is an analysis of why or why not something worked well, troubleshooting, and a hyperparameter optimization procedure summary.

```python
# Define hyperparameters configurations to experiment with
hyperparameters = [
    {'learning_rate': 0.0002, 'batch_size': 32, 'latent_dim': 100},
    {'learning_rate': 0.0001, 'batch_size': 64, 'latent_dim': 150},
    {'learning_rate': 0.0003, 'batch_size': 16, 'latent_dim': 80}
]
```

```python
best_hyperparameters = None
best_discriminator_loss = float('inf')
best_generator_loss = float('inf')
best_discriminator_accuracy = 0
best_discriminator_losses = []
best_generator_losses = []
best_discriminator_accuracies = []
```

```python
# Train GAN models with different hyperparameters
for i, params in enumerate(hyperparameters):
    print(f"Training GAN Model {i+1} with Hyperparameters: {params}")
    gan_model = build_gan_model(**params)
    
    discriminator_losses = []
    discriminator_accuracies = []  # List to store discriminator accuracies
    generator_losses = []
    
    # Load and preprocess Monet dataset
    monet_dataset = load_and_preprocess_dataset(MONET_FILENAMES, batch_size=params['batch_size'])
    epochs = 5
    steps_per_epoch = monet_size // params['batch_size']
    monet_dataset = monet_dataset.repeat()  
    for epoch in range(epochs):
        print(f"Epoch {epoch+1}/{epochs}")
        for step, real_images in enumerate(monet_dataset.take(steps_per_epoch)):
            noise = tf.random.normal(shape=(params['batch_size'], params['latent_dim']))  # Ensure noise shape matches latent_dim

            # Generate fake images with the generator
            generated_images = gan_model.layers[1].predict(noise)

            # Combine real and fake images
            x_combined = tf.concat([real_images, generated_images], axis=0)

            # Labels for real and fake images
            y_real = tf.ones((params['batch_size'], 1))
            y_fake = tf.zeros((params['batch_size'], 1))
            y_combined = tf.concat([y_real, y_fake], axis=0)

            # Train discriminator
            d_loss, d_accuracy = gan_model.layers[2].train_on_batch(x_combined, y_combined)  
            discriminator_losses.append(d_loss)
            discriminator_accuracies.append(d_accuracy) 

            # Train generator (discriminator weights are frozen)
            noise = tf.random.normal(shape=(params['batch_size'], params['latent_dim']))
            y_gen = tf.ones((params['batch_size'], 1))
            g_loss = gan_model.train_on_batch(noise, y_gen)
            generator_losses.append(g_loss)

            # Print progress
            print(f"Step {step+1}/{steps_per_epoch}, Discriminator Loss: {d_loss}, Discriminator Accuracy: {d_accuracy}, Generator Loss: {g_loss}")

    avg_discriminator_loss = np.mean(discriminator_losses)
    avg_discriminator_accuracy = np.mean(discriminator_accuracies) 
    avg_generator_loss = np.mean(generator_losses)
    
    # Update best hyperparameters and results if necessary
    if avg_discriminator_loss < best_discriminator_loss and avg_generator_loss < best_generator_loss and avg_discriminator_accuracy > best_discriminator_accuracy:
        best_hyperparameters = params
        best_discriminator_loss = avg_discriminator_loss
        best_discriminator_accuracy = avg_discriminator_accuracy 
        best_generator_loss = avg_generator_loss
        best_discriminator_losses = discriminator_losses
        best_generator_losses = generator_losses
        best_discriminator_accuracies = discriminator_accuracies

print("Best Hyperparameters:", best_hyperparameters)
print("Best Discriminator Loss:", best_discriminator_loss)
print("Best Discriminator Accuracy:", best_discriminator_accuracy)  
print("Best Generator Loss:", best_generator_loss)
```

```python
# Plot losses and discriminator accuracy
plt.figure(figsize=(10, 5))
plt.plot(best_discriminator_losses, label='Discriminator Loss')
plt.plot(best_generator_losses, label='Generator Loss')
plt.plot(best_discriminator_accuracies, label='Discriminator Accuracy')
plt.xlabel('Iteration')
plt.ylabel('Loss / Accuracy')
plt.title('Training Losses and Discriminator Accuracy of Best Model')
plt.legend()
plt.show()
```

```python
# Function to generate Monet-style images using the best hyperparameters
def generate_monet_images(hyperparameters, output_folder):
    # Build GAN model with best hyperparameters
    gan_model = build_gan_model(**hyperparameters)

    # Generate and save Monet-style images
    num_images = np.random.randint(7000, 10001)  # Randomly choose the number of images to generate
    noise = tf.random.normal(shape=(num_images, hyperparameters['latent_dim']))  # Generate random noise
    generated_images = gan_model.layers[1].predict(noise)  # Generate images with the generator

    # Create output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Save generated images as jpg files
    for i, image in enumerate(generated_images):
        filename = os.path.join(output_folder, f"monet_image_{i}.jpg")
        tf.keras.preprocessing.image.save_img(filename, image)

    # Zip the generated images
    with zipfile.ZipFile('images.zip', 'w') as zipf:
        for root, dirs, files in os.walk(output_folder):
            for file in files:
                zipf.write(os.path.join(root, file), os.path.relpath(os.path.join(root, file), output_folder))

# Call the function to generate Monet-style images and save them in a zip file
generate_monet_images(best_hyperparameters, './generated_images/')
```
